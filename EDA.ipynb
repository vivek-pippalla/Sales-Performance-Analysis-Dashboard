```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Profit Margin = Profit / Sales
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
# Replace infinite or missing values (like if Sales was 0)
df['Profit Margin'] = df['Profit Margin'].replace([float('inf'), -float('inf')], 0)
df['Profit Margin'] = df['Profit Margin'].fillna(0)
```

```python
print(df['Profit Margin'].isnull().sum())
```

```python
print(df[['Sales', 'Profit', 'Profit Margin']].max())
print(df[['Sales', 'Profit', 'Profit Margin']].min())
```

```python

# List of numeric columns to clean
numeric_cols = ['Sales', 'Profit', 'Profit Margin']

# Clean and convert each column
for col in numeric_cols:
    # Remove commas (e.g., "1,234.56" â†’ "1234.56")
    df[col] = df[col].astype(str).str.replace(',', '').astype(float)
```

```python
import pandas as pd
import re
from decimal import Decimal, InvalidOperation

def clean_money(val):
    try:
        val = re.sub(r'[^\d\-.]', '', str(val))  # remove everything except digits, dot, minus
        return round(Decimal(val), 2)
    except InvalidOperation:
        return 0

df = pd.read_csv("cleaned_superstore_data.csv")

# Strip column names
df.columns = df.columns.str.strip()

# Clean numerical columns
for col in ['Sales', 'Profit']:
    df[col] = df[col].apply(clean_money)

# Profit Margin: float with 4 decimal places
df['Profit Margin'] = pd.to_numeric(df['Profit Margin'], errors='coerce').round(4).fillna(0)

# Dates formatting
for d in ['Order Date', 'Ship Date']:
    df[d] = pd.to_datetime(df[d], errors='coerce').dt.strftime('%Y-%m-%d')

# Save final cleaned file
df.to_csv("cleaned_for_mysql.csv", index=False, encoding='utf-8')
```

# Monthly sales over time


```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales', figsize=(12,6))
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```

# Profit by Region

```python
plt.figure(figsize=(9,6))
sns.barplot(data=df, x='Region', y='Profit')
plt.title('Profit by Region')
plt.show()
```

# Top products by sales


```python
top_products_sales = df.groupby('Product Name').sum(numeric_only=True).sort_values('Sales', ascending=False).head(10)
top_products_sales['Sales'].plot(kind='barh', title='Top 10 Products by Sales', figsize=(10,6))
plt.gca().invert_yaxis()
plt.xlabel('Sales')
plt.show()
```

# Top products by profit

```python

top_products_profit = df.groupby('Product Name').sum(numeric_only=True).sort_values('Profit', ascending=False).head(10)
top_products_profit['Profit'].plot(kind='barh', title='Top 10 Products by Profit', figsize=(10,6), color='green')
plt.gca().invert_yaxis()
plt.xlabel('Profit')
plt.show()
```

# Sales and Profit by Category


```python
category_perf = df.groupby('Category').sum(numeric_only=True)[['Sales', 'Profit']]
category_perf.plot(kind='bar', title='Sales and Profit by Category', figsize=(8,5))
plt.ylabel('Amount')
plt.show()
```

# Sales and Profit by Sub-Category


```python
plt.figure(figsize=(12,6))
subcat_perf = df.groupby('Sub-Category').sum(numeric_only=True)[['Sales', 'Profit']]
subcat_perf.sort_values('Sales', ascending=False).plot(kind='bar', title='Sales and Profit by Sub-Category')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show()
```

# Quantity vs Profit (Is bulk buying less profitable?

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Quantity', y='Profit', hue='Category')
plt.title('Quantity vs Profit')
plt.show()
```

# Impact of Discounts on Profit

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='Discount', y='Profit', hue='Category')
plt.title('Discount vs Profit')
plt.show()
```

# State-wise Analysis

```python
# Top 10 States by Loss
loss_states = df.groupby('State').sum(numeric_only=True).sort_values('Profit').head(10)
plt.figure(figsize=(10, 6))
loss_states['Profit'].plot(kind='barh', color='red')
plt.title('Top 10 States by Loss')
plt.xlabel('Profit')
plt.gca().invert_yaxis()
plt.show()
```
